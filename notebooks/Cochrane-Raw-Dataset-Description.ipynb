```python
import json
import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from os import path
from nltk import sent_tokenize, word_tokenize
from statistics import mean
```

```python
# Config
sns.set_theme(style="white", context="talk")
sns.set_context('notebook')
base_dir = "../src/data/cochrane/scraped_data"
```

```python
with open(path.join(base_dir, "data.json"), "r", encoding="utf-8") as f:
    json_data = json.load(f)
```

```python
len(json_data)
```

```python
data = {"lang":[]}
for review in json_data:
    for lang in review["content"]:
        data["lang"].append(lang)
df = pd.DataFrame(data)
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax = sns.countplot(x=df["lang"], ax=ax)

ax.set(xlabel='Languages', ylabel='Count')
for i in ax.containers:
    ax.bar_label(i,)
sns.despine(bottom = True, left = True)
plt.show()
plt.savefig("04-cochrane-raw-language-distribution.png")
```
